### Libraries ###
The following project eases the task of sentiment analysis by using the python library textblob which indeed uses NLTK and pattern libraries in its backend, in the above cell we import textblob. For more, info on textblob docs check this [link](https://textblob.readthedocs.io/en/dev/)
In the first part of this project, to perform sentiment analysis, we use the dataset of shakespeare's macbeth which comes preinstalled with the corpora of [NLTK](https://github.com/nltk/nltk)

```python
print (text)
```

```python
blob=TextBlob(text)
```

In the above cell we read the .txt file of macbeth and store it in a variable . It is later passed on as a parameter to the TextBlob function ,so that we can perform certain functions present in the textbolb library on the text file.It is later stored in a variable called blob.
Beneath, we are using the parts of speech tagging function and checking the tags provided to the words in macbeth, i.e preposition, adjective, adverb , verb ,etc.

```python
blob.tags
```

We check out the nouns present in macbeth dataset and print them using textblob's noun_phrases method

```python
blob.noun_phrases
```

We initialize three variables as zeros, i - counter variable keeps a count on the total no of sentences in macbeth, pos- keeps a count on the number of sentences with positive polarity in macbeth, neg- keeps a count on the number of negative sentences in the document.

```python
i=0
pos=0
neg=0
for sentence in blob.sentences:
    print (sentence.sentiment.polarity)
    i+=1
    if sentence.sentiment.polarity>0:
        pos+=1
    elif sentence.sentiment.polarity<0:
        neg+=1
        
print ('No of sentences:{}'.format(i))
print ('No of positive sentences:{}'.format(pos))
print ('No of negative sentences:{}'.format(neg))
```
